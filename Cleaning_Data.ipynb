```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/AlphaCurse/DATA607Project3/main/multipleChoiceResponses.csv', encoding='latin-1')
df
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
df2 = df[['GenderSelect','Country','Age','EmploymentStatus','CurrentJobTitleSelect','TitleFit']]
df2
```

```python
df2['GenderSelect'].value_counts()
```

```python
df2['GenderSelect'].fillna(df['GenderSelect'].mode()[0], inplace=True)
df2['Country'].replace("People 's Republic of China","People's Republic of China", inplace=True)
df2['Country'].fillna('Other', inplace=True)
df2['Age'].fillna(df.Age.median(), inplace=True)
df2['CurrentJobTitleSelect'].fillna('Other', inplace=True)
df2['TitleFit'].fillna(df['TitleFit'].mode()[0], inplace=True)
```
